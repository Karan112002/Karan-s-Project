### 4. Compute the basic numerical aggregates of the male and female weights (measures of location, dispersion, and shape). In your own words, describe and compare the two distributions (e.g., are they left skewed, which one has more dispersion, and so forth). 

```python
import numpy as np
from scipy.stats import skew, kurtosis

# Define a function to compute aggregates
def summarize_weights(weights):
    summary = {
        # Measures of Location (Central Tendency)
        "Count": len(weights),
        "Mean": weights.mean(),
        "Median": np.median(weights, axis=0),

        # Measures of Dispersion (Spread)
        "Standard Deviation": weights.std(),
        "variances": weights.var(),
        "Range": weights.max() - weights.min(),
        "IQR": np.percentile(weights, 75) - np.percentile(weights, 25),
        "MAD": np.mean(np.abs(weights - np.mean(weights))),

        # Measures of Shape (Distribution)
        "Skewness": skew(weights, bias=False),
        "Kurtosis": kurtosis(weights, bias=False)
    }
    return summary

# Get summaries
female_summary = summarize_weights(female_weights)
male_summary = summarize_weights(male_weights)

# Convert to DataFrame for better display
summary_df = pd.DataFrame([female_summary, male_summary], index=["Female", "Male"])
print(summary_df)
```

## Summary:
This code calculates and displays the following stats for both male and female weight data:

Count

Mean & Median

Standard Deviation, Variance, and Range, Mean absolute error, IQR

Skewness (asymmetry of the distribution)

Kurtosis (tailedness of the distribution)

The result is a clean, tabular summary that lets you quickly compare how male and female weights are distributed.

### 10. In your own words, list some advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio. 

## 1. Body Mass Index (BMI)

**Advantages:**

1. Simple and quick: Easy to calculate using only weight and height.

2. Widely used: Recognized globally for screening weight categories.

3. Standardized: Allows easy comparison across populations.

**Disadvantages:**

1. Doesn't account for muscle mass: A muscular person might be classified as overweight.

2. Ignores fat distribution: Two people with the same BMI can have very different health risks.

3. Not tailored to individuals: Doesn’t consider age, sex, or ethnicity well.

## 2. Waist-to-Height Ratio (WHtR)

**Advantages:**

1. Better risk indicator: More directly linked to cardiometabolic risk (e.g., heart disease, diabetes).

2. Includes fat distribution: Specifically central (abdominal) fat, which is more dangerous.

3. One-size-fits-all cutoff: A value > 0.5 is generally a red flag across age and sex.

**Disadvantages:**

1. Less familiar to the public: Not as widely used as BMI in health screening.

2. Still doesn’t differentiate between fat and muscle.

## 3. Waist-to-Hip Ratio (WHR)

**Advantages:**

1. Highlights fat distribution: Particularly helpful in identifying “apple-shaped” vs. “pear-shaped” bodies.

2. Gender-specific insight: Can reflect different health risks for men vs. women.

**Disadvantages:**

1. Measurement sensitive: Small errors in waist or hip measurements affect the result.

2. Doesn’t reflect overall body size: A person with a high WHR might still be underweight or overweight overall.

3. Interpretation depends on sex and population: Different healthy ranges for men and women.


### 11. Print out the standardised body measurements for the 5 persons with the lowest BMI and the 5 persons with the 5 highest BMI (e.g., call print for a subset of zfemale comprised of 10 chosen rows as determined by a call to numpy.argsort). Interpret the results. 

```python
import pandas as pd
import numpy as np

# Load data
female_df = pd.read_csv("nhanes_adult_female_bmx_2020.csv", comment="#")

# Convert DataFrame to NumPy matrix
female = female_df.to_numpy()

# Compute BMI: weight (col 0) / (height (col 1) in meters)^2
female_bmi = female[:, 0] / (female[:, 1] / 100) ** 2

# Add BMI as the last column
female_with_bmi = np.column_stack((female, female_bmi))

# Select columns of interest: weight, height, hip, waist, BMI
# Make sure these indices match the column order in your dataset
columns_of_interest = [0, 1, 5, 6, 7]  # 7 is BMI which we just added
female_subset = female_with_bmi[:, columns_of_interest]

# Standardize
zfemale = (female_subset - np.mean(female_subset, axis=0)) / np.std(female_subset, axis=0)

# Get indices of 5 lowest and 5 highest BMI
bmi_column = female_subset[:, -1]  # last column is BMI
bmi_sorted_indices = np.argsort(bmi_column)
lowest_5 = bmi_sorted_indices[:5]
highest_5 = bmi_sorted_indices[-5:]
selected_indices = np.concatenate([lowest_5, highest_5])

# Get the standardized values for those individuals
zfemale_subset = zfemale[selected_indices]

# Convert to DataFrame for labeling columns
col_names = ["Weight", "Height", "Hip", "Waist", "BMI"]
zfemale_df = pd.DataFrame(zfemale_subset, columns=col_names)

# Print
print("Standardized body measurements for 5 lowest and 5 highest BMI (Females):\n")
print(zfemale_df)
```

## Summary
- This code calculates the BMI for each individual in the dataset.

- It standardizes the relevant measurements (Weight, Height, Hip, Waist, BMI).

- Then, it identifies the 5 individuals with the lowest and highest BMI values.

- Finally, it displays the standardized measurements of these selected individuals in a pandas DataFrame.

This analysis can be helpful for identifying outliers in the dataset and observing how body measurements relate to BMI across different individuals.
